```python
from hybridtablerag.embeddings.embedding_store import EmbeddingStore
import pandas as pd

sales = pd.read_csv("../data/sales.csv")
sales_rows = ["|".join(map(str, row)) for row in sales.values]
sales_meta = [{"table": "sales", "row_idx": i} for i in range(len(sales_rows))]

store = EmbeddingStore()
store.build_index(sales_rows, sales_meta)

results = store.search("High quantity orders in March 2023", top_k=3)
for r in results:
    print(r)
```

```python
from hybridtablerag.retrieval.semantic_retriever import SemanticRetriever

rag = SemanticRetriever(
    llm_name="gemini",
    llm_model="gemini-2.5-flash",
    embedding_store=store   # now supported
)

query = "High quantity orders in March 2023"
answer = rag.answer_query(query, top_k=3)
print("Answer:", answer)
```
